```python
import _pickle as pickle
import omniglot
import numpy as np
```

```python
#=====================================================#
#使用データの選択
#　　small1 -> Trainデータ:2040   Testデータ数:680     総クラス数:135
#　　small2 -> Trainデータ: 2340  Testデータ数: 780   総クラス数:155
#　　large　　 -> Trainデータ:14460 Testデータ数:4820  総クラス数:693
#=====================================================#
dataname = "small1"
```

```python
#データの取得
dataset = np.load("DATA/{}.npz".format(dataname))
train_data = dataset["train_data"]
train_label = dataset["train_label"]
test_data = dataset["test_data"]
test_label = dataset["test_label"]
```

```python
#=====================================================#
#問題セットの作成
#  QUESTION_NUM : 問題数
#  CHOICE_NUM : １問あたりの選択肢数
#
#  get_omniglot_questionsetは問題数の大きさのリストLを返します。
#  L[i]は"question_data","question_label","ans_data","ans_label"をキーに持つ辞書です。
#  "question_data" : 問題の画像 (1,1,105,105)
#  "question_label" : 問題のラベル (1)
#  "ans_data" : 選択肢の画像 (CHOICE_NUM,1,105,105)
#  "ans_label" : 選択肢の画像 (CHOICE_NUM)
#=====================================================#
train_questionset = omniglot.get_omniglot_questionset(train_data,train_label,QUESTION_NUM=1000, CHOICE_NUM = 5)
test_questionset = omniglot.get_omniglot_questionset(test_data,test_label,QUESTION_NUM=300, CHOICE_NUM = 5)
print(len(train_questionset))
train_questionset[0]
```

```python
#=============================#
#問題セットの確認
#  line : 表示する問題数
#
#  Questiion,Choiceの横の数字はクラス番号
#=============================#
%matplotlib inline
omniglot.show_questionset(train_questionset,line=7)
```
